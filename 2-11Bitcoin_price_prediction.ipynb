```python
df[df["Weighted_Price"].isna()]
```

```python
#handling na values by using forward fill

df["Int_Weighted_Price"] = df["Weighted_Price"].ffill()
```

```python
df.Open = df.Open.ffill()
```

```python
df["Int_Weighted_Price"]
```

```python
df.Close = df.Close.ffill()
```

```python
df["Int_Weighted_Price"]
```

```python
# We noted that NA value was filled with the same number if we use forward filling function. We use rolling mean for window 4 to average those number instead.
df["New_Weighted_Price"] = df["Int_Weighted_Price"].rolling(window=4).mean()
```

```python
df
```

```python
# We noted that when we filled NA number with average by rolling mean window 4 the other value of effected that should be correct since the we just want to fill NA value and the other value remain the same,so we use below funtion to keep value the same by just filling NA columns.
import numpy as np

df["Cln_Weighted_Price"] = np.where(df["Weighted_Price"].isnull(),df["New_Weighted_Price"],df["Weighted_Price"])
```

```python
df
```

```python
#Now we've got new cleaned data below
df["Cln_Weighted_Price"]
```

# Decomposition

```python
plt.figure(figsize=[20,10])
decomp_viz = sm.tsa.seasonal_decompose(df_daily.Cln_Weighted_Price)
fig = decomp_viz.plot()
fig.set_size_inches((20, 10))
# Tight layout to realign things
fig.tight_layout()
plt.show()
```

```python
decomp_viz.resid
```

```python
# Testing stationary by Adfuller test
sm.tsa.stattools.adfuller(decomp_viz.resid.dropna())
#PDQ -- reverse engineer
```

# MODELLING

```python
decomp_viz.resid.dropna()
```

```python
# Using Auto Arima to run the model

from pmdarima.arima import auto_arima
```

```python
model = auto_arima(decomp_viz.resid.dropna(), trace=True, error_action='ignore', suppress_warnings=True, seasonal=True, m=7)
```

```python
tr_model = model.fit(decomp_viz.resid.dropna())
```

```python
train = decomp_viz.resid.dropna()
```

```python
train
```

```python
predict = tr_model.predict(n_periods=50)
```

```python
plt.figure(figsize=(20,10))

#plot the predictions for validation set
plt.plot(train.tail(100), label='Train')

#plt.plot(valid, label='Valid')

plt.plot(predict, label='Prediction')


plt.show()
```
